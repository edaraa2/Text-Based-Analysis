# USING REGEX

```python
import regex as re # importing and using regex for  matching and extracting string patterns

def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
```

# TESTTOKENS

```python
testtext = "The Finance company name called cowbell, which is located at SF Bay Area laid off 28 people in year 2015"
testtokens = tokenize(testtext) # splitting up a larger body of text into smaller lines, words or even creating words for a non-English language.
 
print(testtokens)
```

# IMPORTING NLTK PACAKAGE

```python
import nltk # importing Natural Language Toolkit for  which helps in tokenization..
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

# STOPWORDS

```python
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]
#Just testing it
cleantokens= remove_stop(testtokens)
print(cleantokens)
```

# TOKENS

```python
pipeline = [str.lower, tokenize, remove_stop]

def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

# USING COLLECTION AND IMPORTING COUNTER

```python
from collections import Counter # importing counter to count variables

tokens = tokenize("Finance company name called cowbell, which is located at SF Bay Area")

testcount= Counter(tokens)
print(testcount)
```

# TOKENIZATION

```python
more_tokens = tokenize("cowbell is located at SF Bay Area")
testcount.update(more_tokens)
print(testcount)
```

# WORDCLOUD

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Reading text from Excel file
df = pd.read_excel("C:/Users/anush/OneDrive/Desktop/Text Based Analysis/Lay-offs.xlsx")

# Assuming 'text_column_name' is the name of the column containing the text data
# Combining text from all rows in the specified column
all_text = ' '.join(df['company'].astype(str))

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

# Plotting word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
